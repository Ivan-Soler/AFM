```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file="../../fractional/gaus_pol/data_fractionals.txt"
```

```python
data=[]
data_file=open(file,"r")
for line in data_file:
    reline=line.replace("(","").replace(")","").replace(",","")
    split=reline.split(" ")
    split.pop();
    #print(split)
    data.append(np.float_(split))
data_file.close()
data=np.array((data))
```

```python
x=data[:,0]
ind2=6+3*5
y=data[:,ind2]
#y=data[:,ind2]
plt.scatter(x,y)
```
