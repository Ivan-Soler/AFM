### Instanton Counting

```python
#From the susy plot
sizes=[4,4,4,32]
directory="../ssd/4x4x4x32/b2p44_new/compare_4p0t/"
measures=["gf_afm_4p0t/", "gf_afm_3p0t/", "gf_afm_2p0t/","gf_afm_1p5t/", 
          "gf_afm_1p125t/", "gf_afm_0p75t/", "gf_afm_0p5t/", "gf_afm_0p25t/","gf_afm_0p0t/"]
instantons={}
mean_instanton=open(directory+"instantons_mean.txt","w")
for measure in measures:
    maxima_file=open(directory+measure+"instantons.txt","w")
    maxima_mean=0
    minima_mean=0
    for file in os.listdir(directory+measure):
        if "c.txt" in file:
            #print(file)
            mode_num=""
            for char in file:
                if char.isdigit():
                    mode_num+=char
            susy_file=open(directory+measure+file)
            data=susy_file.read().split("\n")
            data.pop()
            data=np.array(data, dtype=float)
            susy_file.close()
            maxima=Maxima.simple_1d(data,sizes[3],0.02)
            minima=Maxima.simple_1d(-data,sizes[3],0.02)
            instantons[mode_num]=[len(maxima),len(minima)]
            #print(mode_num)
            print(mode_num,len(maxima),len(minima),file=maxima_file)
            maxima_mean+=len(maxima)
            minima_mean+=len(minima)
    maxima_file.close()
    print(measure,maxima_mean/100,minima_mean/100,(maxima_mean+minima_mean)/100, (maxima_mean-minima_mean)/100,file=mean_instanton)
mean_instanton.close()          
```

### Xi max and GF

```python
def Xi_max(data,susy_max):
    maximum=0
    for i in range(0,len(data[0])):
            if (data[0,i] > maximum) and (data[1,i]<susy_max[0]):
                maximum=data[0,i]
    return(maximum)
```

```python
def find_max(folder,measure,observable):
    data=np.loadtxt(folder+measure+observable+".txt")
    if measure=="gf_afm_0p0t/":
        susy_max=np.loadtxt(folder+measure+"end_spectrum.txt")
    else:
        susy_max=np.loadtxt(folder+measure+"end_spectrum.txt")
    maximum=Xi_max(data,susy_max)
    return(maximum)
```

### Checking plaquettes

```python
sizes=[4,4,4,32]
temp_file="../check_plaquettes/Plaquettes.dat"
plaquettes = array.array('d')
planes=6
with open(temp_file, 'rb') as file:
    for i in range(0,10):
        data=int.from_bytes(file.read(4),byteorder='little')
        print(data)
    n=sizes[0]*sizes[1]*sizes[2]*sizes[3]*planes
    plaquettes.fromfile(file,n)
 
```

```python
position=[]
for t in range(0,sizes[3]):
    for z in range (0,sizes[2]):
        for y in range (0,sizes[1]):    
            for x in range (0,sizes[0]):
                for plane_i in range(0,planes):
                    if ((x==sizes[0]-1) and (y==sizes[1]-1) and (plane_i==0)) or ((y==sizes[1]-1) and (z==sizes[2]-1) and (plane_i==2)) or ((z==sizes[2]-1) and (x==sizes[0]-1) and (plane_i==1)):
                        position.append(plane_i+planes*(x+sizes[0]*(y+sizes[1]*(z+sizes[2]*t))))
```
