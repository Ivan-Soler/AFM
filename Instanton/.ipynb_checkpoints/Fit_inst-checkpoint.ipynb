```python
def inst_plot(position,maxima_x,maxima_y,rho):
    #maxim=[23,23]
    return(3/(np.pi*rho**4)*(rho**2/((position[0]-maxima_x)**2
                        + (position[1]-maxima_y)**2+rho**2))**4)
```
